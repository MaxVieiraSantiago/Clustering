```python
import pandas as pd

# pacote para geração de gráficos
import matplotlib.pyplot as plt

# pacote para criação de regressão, pontos, ( x, y )
from sklearn.datasets import make_regression

# pacote para Modelo de Regressão Linear
from sklearn.linear_model import LinearRegression
```

```python
x, y = make_regression( n_samples=200, n_features=1, noise=30 )

plt.scatter( x, y )
```

```python
# instanciamento do modelo
modelo = LinearRegression()

# treinamento do modelo utilizando os valores gerados por make_regression ( 200 amostras, 1 feature, ruído = 30)
modelo.fit( x, y )
```

```python
coeficiente_angular = modelo.coef_
coeficiente_linear  = modelo.intercept_

print( "Coeficiente Angular: {0},   Coeficiente Linear: {1}".format( coeficiente_angular, coeficiente_linear ) )

# y = a + bx  , onde ( a = coeficiente Linear, b = coeficiente Angular, x são os valores de x )
# y = coeficiente_linear + coeficiente_angular * x
```

```python
# Valores iniciais de ( x, y )
plt.scatter( x, y )

# Geração da Reta utilizando os coeficientes encontrados
plt.plot( x, coeficiente_linear + coeficiente_angular * x, color='red')
plt.show()
```

```python
# Valores iniciais de ( x, y )
plt.scatter( x, y )

# Geração da Reta utilizando os coeficientes encontrados
plt.plot( x, coeficiente_linear + coeficiente_angular * x, color='red')

# Validação de valor para x
x_aux = 1.5
plt.scatter( x_aux, coeficiente_linear + coeficiente_angular * x_aux, color='green', s=50)

plt.show()
```
